```python
# import block
import os, sys
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data_set = pd.read_csv('./data/IoTProcessed_Data.csv')
data_set
```

```python
# clean up and adjusting 

data_set_clean = data_set.copy()
data_set_clean["date"] = pd.to_datetime(data_set_clean["date"])
data_set_clean = data_set_clean.sort_values(by='date')
# Rename the column temporarily for clarity
data_set_clean = data_set_clean.rename(columns={"tempreature": "temperature"})
data_set_clean
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Plot Date vs Temperature
plt.figure(figsize=(10, 6))
plt.plot(data_set_clean['date'], data_set_clean['temperature'], label='Temperature', linewidth=2)

# Add labels and title
plt.xlabel('Date')
plt.ylabel('Temperature (Â°C)')
plt.title('Temperature Over Time')
plt.legend()
plt.grid(True)
plt.show()
```
